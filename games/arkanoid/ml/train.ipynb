```python
import pickle
Ball_x = []
Ball_y = []
Ball_speed_x = []
Ball_speed_y = []
Dest = []
Platform = []
Command = []
file_list = ["random_successe4(1)","random_successe4(2)","random_successe4(3)","random_successn3(1)","random_successn3(2)"
             ,"random_successn4(1)","random_successn4(2)","random_successn4(4)","random_successn5(1)","random_successn5(2)",
            "random_successn5(3)","random_successn5(4)","random_successn5(5)","random_successn5(6)","random_successn5(7)",
            "random_successn5(8)","random_successe4(4)"]
for j in file_list:
    path = "C:/master/games/arkanoid/log/"+j+".pickle"
    file = open(path, "rb")
    temp_scene_info = []
    temp_command = []
    data = pickle.load(file)
    temp_scene_info = temp_scene_info + data['scene_info']
    temp_command = temp_command + data['command']
    file.close()
    index=[]
    for i, s in enumerate(temp_scene_info[1:-2]):
        x_ball = s['ball'][0]
        y_ball = s['ball'][1]
        plat = s['platform'][0]
        x_speed = temp_scene_info[i+2]["ball"][0] - temp_scene_info[i+1]["ball"][0]
        y_speed = temp_scene_info[i+2]["ball"][1] - temp_scene_info[i+1]["ball"][1]
        if x_speed ==0 or y_speed == 0:
            index.append(i)
            continue
        Ball_x.append(x_ball)
        Ball_y.append(y_ball)
        Platform.append(plat)
        Ball_speed_x.append(x_speed)
        Ball_speed_y.append(y_speed)
        dest = 100
        if y_speed != 0:
            slope=x_speed/y_speed
        if y_ball > 320 and y_speed <0:
            dest = x_ball
        elif y_speed > 0 and y_ball > 270 :
            dest=x_ball+(395-y_ball)*slope
            if dest >= 390:
                dest=dest-390
            elif dest >=195:
                dest=390-dest
            elif dest <=0 and dest >=-195:
                dest=dest*(-1)
            elif dest < -195:
                dest =dest +390
        Dest.append(dest)          
    for i in range(len(temp_command[1:-2])):
        if i in index:
            continue
        elif temp_command[1:-2][i] == "NONE":
            Command.append(0)
        elif temp_command[1:-2][i] == "MOVE_LEFT":
            Command.append(-1)
        elif temp_command[1:-2][i] == "MOVE_RIGHT":
            Command.append(1)
        else:
            Command.append(0)
```

```python
import numpy as np

numpy_data = np.array([Ball_x, Ball_y, Ball_speed_x, Ball_speed_y, Dest, Platform])
X = np.transpose(numpy_data) 
y = Command
```

```python
X, y
```

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=3)
print(model.fit(X, y))
print(model.score(X, y))
```

```python
with open('my_model.pickle', 'wb') as f:
    pickle.dump(model, f)
```
